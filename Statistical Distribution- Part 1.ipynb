```python
#Import Required Libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, norm, nbinom, hypergeom
```

```python
# Poisson Distribution
lambda_param = 3
poisson_data = np.random.poisson(lambda_param, 1000)

plt.hist(poisson_data, bins=30, density=True, alpha=0.6, color='b')
plt.title('Poisson Distribution')
plt.show()
```

```python
# Normal Distribution
mu, sigma = 0, 1
normal_data = np.random.normal(mu, sigma, 1000)

plt.hist(normal_data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Normal Distribution')
plt.show()
```

```python
# Negative Binomial Distribution
n = 5
p = 0.3
negative_binomial_data = np.random.negative_binomial(n, p, 1000)

plt.hist(negative_binomial_data, bins=30, density=True, alpha=0.6, color='r')
plt.title('Negative Binomial Distribution')
plt.show()
```

```python
# Hypergeometric Distribution
M = 20
n = 7
N = 12
hypergeom_data = np.random.hypergeometric(M, n, N, 1000)

plt.hist(hypergeom_data, bins=30, density=True, alpha=0.6, color='y')
plt.title('Hypergeometric Distribution')
plt.show()
```
