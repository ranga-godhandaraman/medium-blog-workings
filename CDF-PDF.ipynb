```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Define parameters
mu = 0  # Mean
sigma = 1  # Standard deviation
```

```python
# Generate random samples from a standard normal distribution
data = np.random.normal(mu, sigma, 1000)
```

```python
# Compute PDF and CDF
pdf = stats.norm.pdf(data, mu, sigma)
cdf = stats.norm.cdf(data, mu, sigma)
```

```python
# Compute moments
mean = np.mean(data)
variance = np.var(data)
skewness = stats.skew(data)
kurtosis = stats.kurtosis(data)
```

```python
# Compute characteristic function
def characteristic_function(t):
    return np.mean(np.exp(1j * t * data))

# Display results
print("Mean:", mean)
print("Variance:", variance)
print("Skewness:", skewness)
print("Kurtosis:", kurtosis)
print("Characteristic function at t=0.5:", characteristic_function(0.5))
```

```python
# Generate data for standard normal distribution
x = np.linspace(-5, 5, 1000)
pdf = stats.norm.pdf(x)
cdf = stats.norm.cdf(x)
```

```python
# Plot PDF
plt.figure(figsize=(10, 5))
plt.plot(x, pdf, label='PDF')
plt.title('Probability Density Function (PDF) of Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plot CDF
plt.figure(figsize=(10, 5))
plt.plot(x, cdf, label='CDF', color='red')
plt.title('Cumulative Distribution Function (CDF) of Standard Normal Distribution')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.legend()
plt.show()
```
