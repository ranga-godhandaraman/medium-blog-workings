```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import t
# Parameters
df = 10  # Degrees of freedom

# Generate random variates
x = np.linspace(-4, 4, 1000)
y = t.pdf(x, df)

# Plotting
plt.plot(x, y, label='T-Distribution (df=10)',color='r')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('T-Distribution')
plt.legend()
plt.show()
```

```python
from scipy.stats import f

# Parameters
dfn = 5  # Degrees of freedom numerator
dfd = 10  # Degrees of freedom denominator

# Generate random variates
x = np.linspace(0.01, 5, 1000)
y = f.pdf(x, dfn, dfd)

# Plotting
plt.plot(x, y, label='F-Distribution (dfn=5, dfd=10)', color='g')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('F-Distribution')
plt.legend()
plt.show()
```

```python
from scipy.stats import chi2

# Parameters
df = 5  # Degrees of freedom

# Generate random variates
x = np.linspace(0, 20, 1000)
y = chi2.pdf(x, df)

# Plotting
plt.plot(x, y, label='Chi-Square Distribution (df=5)',color='b')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Chi-Square Distribution')
plt.legend()
plt.show()
```
